```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
filename = '/content/cereal.xlsx'
df = pd.read_excel(filename)
df.head()
```

```python
df_filter_type_cold = df["type"]== "Cold"
df_t_Cold = df[df_filter_type_cold]
```

```python
df_t_Hot = df[~df_filter_type_cold]
```

```python
df_t_Cold["type"].hist(alpha=0.75,edgecolor = "black",label = 'Dog')
plt.ylabel("Count")
plt.xlabel("Type Cereal")

plt.style.use('seaborn')
df_t_Hot["type"].hist(facecolor='orange',alpha=0.75,edgecolor = "black",label = 'Cat')
#plt.set_title("Cantidad de Tipo de Cereal preferidos por los clientes")
plt.legend()
```

```python
mean_calories_manufacturer = df.groupby(['type','Manufacturer'])[['calories per serving']].mean().round(2).reset_index()
mean_calories_manufacturer = mean_calories_manufacturer.rename(columns={'calories per serving':'Mean_calories/serving(gr)'})
mean_calories_manufacturer.columns
```

```python


plt.figure(figsize = (10,5))
plt.xticks(rotation=45)
sns.barplot(x = 'Manufacturer', y = 'Mean_calories/serving(gr)',data= mean_calories_manufacturer,hue='type',edgecolor='b',color='y')
```

```python
mean_sugar_manufacturer = df.groupby(['type','Manufacturer'])[['grams of sugars']].mean().round(2).reset_index()
mean_sugar_manufacturer = mean_sugar_manufacturer.rename(columns={'grams of sugars':'grams of sugars'})
mean_sugar_manufacturer
```

```python
plt.figure(figsize = (10,5))
plt.xticks(rotation=45)
sns.barplot(x = 'Manufacturer', y = 'grams of sugars',data= mean_sugar_manufacturer,hue='type',edgecolor='b',color='g')
```
